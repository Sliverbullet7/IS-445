```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import bqplot.pyplot as bqplt
from bqplot import *
import pandas as pd
```

```python
raw_data = pd.read_csv("https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv")
raw_data.head()
                       
```

```python
filt_data = raw_data[raw_data['Year Acquired'] != 0]
sort_data = filt_data.sort_values(by='Year Acquired')
congress_dis = np.sort(list(set(raw_data['Congress Dist'])))
agency_name = np.sort(list(set(raw_data['Agency Name'])))
sort_data = sort_data[['Agency Name', 'Congress Dist',
                       'Year Acquired', 'Square Footage']]
sort_data.head()
```

```python
square_footage = np.zeros((len(agency_name), len(congress_dis)))
for i, name in enumerate(agency_name):
    for j, dis in enumerate(congress_dis):
        dis = np.around(dis, 2)
        temp_data = sort_data[sort_data['Congress Dist'] == dis][sort_data['Agency Name'] == name]
        square_footage[i][j] = np.sum(temp_data['Square Footage'])
```

```python
from IPython.display import display
from ipywidgets import Output

fig = bqplt.figure()
grid_map = bqplt.gridheatmap(
    square_footage,
    row=agency_name,
    column=congress_dis,
    interactions={"click": "select"},
    selected_style={"stroke": "Red", "stroke-width": 3}
)

out = Output()
@out.capture()
def print_event(self, target):
    print(target)

grid_map.on_element_click(print_event)
display(fig)
# display(figure)
display(out)
```

```python
from IPython.display import display
from ipywidgets import Output
import ipywidgets as widgets

size = 100
scale = 100.
np.random.seed(0)
x_data = np.arange(size)
y_data = np.cumsum(np.random.randn(size) * scale)

x_sc = LinearScale()
y_sc = LinearScale()

ax_x = Axis(label='X', scale=x_sc, grid_lines='solid')
ax_y = Axis(label='Y', scale=y_sc, orientation='vertical', grid_lines='solid')

fig_layout = widgets.Layout(width='100', height='100')

fig = []

out = Output()
@out.capture()
def print_event(self, target):
    print(target)


@out.capture()
def plot_event(self, target):
    factor = 0.5
    # bqplt.clear()
    bqplt.close(fig[1])
    line = Lines(x=x_data, y=x_data*0.5, scales={'x': x_sc, 'y': y_sc})
    fig[1] = Figure(axes=[ax_x, ax_y], marks=[line], title='Square Footage per year',
            fig_margin=dict(top=30, bottom=10, left=20, right=20), layout=fig_layout)
    display(widgets.HBox(fig))
    # display(out)

grid_map = bqplt.gridheatmap(
    color=square_footage,
    row=agency_name,
    column=congress_dis,
    interactions={"click": "select"},
    selected_style={"stroke": "Red", "stroke-width": 3}
)
grid_map.on_element_click(plot_event)

fig.append(Figure(marks=[grid_map], title='Heatmap',
           fig_margin=dict(top=30, bottom=10, left=20, right=20), layout=fig_layout))


line = Lines(x=x_data, y=x_data, scales={'x': x_sc, 'y': y_sc})
fig.append(Figure(axes=[ax_x, ax_y], marks=[line], title='Square Footage per year', fig_margin=dict(top=30, bottom=10, left=20, right=20), layout=fig_layout))

display(widgets.HBox(fig))
display(out)
```
